```python
import findspark
findspark.init()
import pyspark
from pyspark.mllib.fpm import FPGrowth
import re
from random import random
from pyspark import SparkContext
```

```python
import os
port = 4040 + hash(os.getcwd().split("/")[2])%200
config = pyspark.SparkConf().set('spark.executor.memory', '512M')\
         .set('spark.cores.max', '16')\
         .set('spark.port.maxRetries','200')\
         .set('spark.ui.port', port)    
print port,     os.getcwd().split("/")[2] 

sc = pyspark.SparkContext(appName="", master='',conf = config )
```

```python

data = sc.textFile("data/wikipedia_access_log_large.log")

print data.take(5)
```

```python
totalPages = data.count()
print 'There were {:d} pages accessed in total.'.format(totalPages)
```

```python
engLink = 'http://en.wikipedia.org/wiki/'
exclusionList = ['.jpg','.png','.js','.css','.gif','title=','monobook','.php','.ico','Special:']

def is_part(some_string, target):
    return target not in some_string

pageDB_exclusions = data.map(lambda x: x.split(' ')[2]) \
                        .filter(lambda x: all(is_part(x, t) for t in exclusionList) \
                                and (x != 'http://en.wikipedia.org/wiki/') )

print 'There were {:d} pages accessed after filtering with exclusions list.'.format(pageDB_exclusions.count())
```

```python
pageDB = pageDB_exclusions.filter(lambda x: engLink in x )
print 'There were {:d} English pages accessed.'.format(pageDB.count())
```

```python
uniquePageDB = pageDB.map(lambda e: [e,1]).reduceByKey(lambda x,y: x+y)
print 'There were {:d} different English pages accessed.'.format(uniquePageDB.count())
```

```python
top40DB = uniquePageDB.takeOrdered(40, lambda v: -v[1])

print '    Wikipedia Page Name \t\t   Times Accessed'
print '------------------------------------------------------------'
i = 1
for page in top40DB:
    print '{:2d}. {:40s} {:8d}'.format(i, page[0].split('/')[-1], page[1])
    i = i + 1
```
