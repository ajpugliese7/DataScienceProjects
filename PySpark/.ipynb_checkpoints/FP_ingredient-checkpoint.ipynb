```python
import findspark
findspark.init()
import pyspark
from pyspark.mllib.fpm import FPGrowth
import re
from random import random
from pyspark import SparkContext
```

```python
import os
port = 4040 + hash(os.getcwd().split("/")[2])%200
config = pyspark.SparkConf().set('spark.executor.memory', '512M')\
         .set('spark.cores.max', '6')\
         .set('spark.port.maxRetries','200')\
         .set('spark.ui.port', port)    
print port,     os.getcwd().split("/")[2]
sc = pyspark.SparkContext(appName="", master='',conf = config )
```

```python
rddData = sc.textFile("recipes.txt")
```

```python
print rddData.take(10)
```

```python
def splitData(line):
    out = []
    d = line.split(';')
    key = "_".join(d[0:2])
    value = d[2:]
    return (key,value)

def keepTheValue(kv):
    out = []
    for v in kv[1]:
        v = ','.join(v)
        out.append(v)
    return out

shopDB = rddData.map(splitData)
shopDB_values = shopDB.groupByKey().map(keepTheValue)
ingredients = shopDB_values.map(lambda x: (x[0].split(','))) \
                                  .map(lambda x: list(set(x)))
    
print shopDB.take(10)
print "\n\n"
print shopDB_values.take(10)
print "\n\n"
print ingredients.take(10)
print "\n\n"
print 'The dataset contains %i different recipes' % (shopDB.groupByKey().count())
```

```python
shopDB_cuisines = shopDB.map(lambda x: (x[0].split('_')[0], 1)) \
                        .reduceByKey(lambda x,y: x + y)

print shopDB_cuisines.sortBy(lambda x: x[1], ascending=False).take(10)
print '\n'
print 'There are %i different cuisines in the dataset' % \
        (shopDB_cuisines.count())
```

```python
ingredients_sorted = ingredients.flatMap(lambda xs: [x for x in xs]) \
                                .map(lambda x: (x, 1)) \
                                .reduceByKey(lambda x,y: x + y) \
                                .sortBy(lambda x: x[1], ascending=False)

print 'TOP-10 ingredients:\n',
for (a,b) in ingredients_sorted.take(10):
    print a,b
    
```

```python
model = FPGrowth.train(ingredients, minSupport=0.05, numPartitions=10)
result = model.freqItemsets().sortBy(lambda x: x[1], False).collect()
print "Ingredients that are frequently used together, for all cuisines:"
for fi in result:
    if len(fi[0]) > 1:
        print(fi)
```

```python
ingredients_mexican = shopDB.filter(lambda x: x[0].split('_')[0] == 'mexican') \
                                         .groupByKey().map(keepTheValue) \
                                         .map(lambda x: (x[0].split(','))) \
                                         .map(lambda x: list(set(x)))
      
    
print "Ingredients that are frequently used together, only Mexican cuisines:"
print '\n'
model = FPGrowth.train(ingredients_mexican, minSupport=0.05, numPartitions=10)
result = model.freqItemsets().sortBy(lambda x: x[1], False).collect()
print "Ingredients that are frequently used together, for all cuisines:"
for fi in result:
    if len(fi[0]) > 1:
        print(fi)

```
