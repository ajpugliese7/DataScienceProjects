# Data

```python
rddData = sc.textFile("/scratch/ISE495/2018_midterm_data/recipes.txt")
# Each line of data represents a recipes, which is like:
# cusine;ID;ingredients1;ingredients2;...
print rddData.take(5)
print "We have:",rddData.count(), "recepies"
```

```python
ingredients_list = rddData.flatMap(lambda x: x.split(';')[2:]) \
                          .map(lambda x: (x, 1)) \
                          .reduceByKey(lambda x, y: x+y) \
                          .map(lambda kv: kv[0]).collect()

ingredients_dic = {}
reverse_ingredients_dic = {}

for i in range(0, len(ingredients_list)):
    ingredients_dic[ingredients_list[i]] = i
    reverse_ingredients_dic[i] = ingredients_list[i]

ingredients_dic_length = len(ingredients_dic)
print 'There are {:d} ingredients in the dataset'.format(ingredients_dic_length)
```

```python
def createSparseVectors(data):
    idx = []
    val = []
    for i in data[1]:
        if ingredients_dic[i] not in idx:
            idx+=[ingredients_dic[i]]
            val+=[float(1)]
    val = np.array(val)/ np.linalg.norm(val) # normalize to unit row
    out = np.stack((idx, val), axis=-1)
    out = sorted(out, key=lambda x: x[0])
    return (data[0], Vectors.sparse(ingredients_dic_length, [i[0] for i in out], [i[1] for i in out]))

def muNormData(data):
    j=0
    muData = []
    dat = data[1].toArray()
    idx = data[1].indices
    for i in idx:
        muData+=[data[1].values[j]-mu[i]]
        j+=1
    return (data[0], Vectors.sparse(ingredients_dic_length, idx, muData))
        
no_recipes = rddData.count()

recipes_rowNorm = rddData.map(lambda x: (x.split(';')[:2], x.split(';')[2:])) \
                 .map(createSparseVectors)

mu = [i/no_recipes for i in recipes_rowNorm.map(lambda kv: kv[1].toArray()).reduce(lambda x,y : x+y)]

recipes_fullNorm = recipes_rowNorm.map(muNormData)
print recipes_fullNorm.take(2)
```

```python
mat = RowMatrix(recipes_fullNorm.map(lambda kv: kv[1]))

svd = mat.computeSVD(3, computeU=True)
U = svd.U       
s = svd.s     
V = svd.V 

print V
```

```python
ital_data = recipes_fullNorm.filter(lambda kv: 'italian'==kv[0][0]).collect()[0:100]
kor_data = recipes_fullNorm.filter(lambda kv: 'korean'==kv[0][0]).collect()[0:100]
test_data = ital_data + kor_data

cuisine_dic = {}
cuisine_list = ['italian', 'korean']
j=0
for i in cuisine_list:
    cuisine_dic[i] = j
    j+=1
```

```python
markers = ['ro','bs']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in test_data:
    f = i[1]
    typ = markers[cuisine_dic[i[0][0]]]
    f = f.dot(V.toArray())
    ax.scatter(f[0], f[1], f[2], c=typ[0], marker=typ[1])
plt.show()
    
```
